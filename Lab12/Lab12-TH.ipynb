### 2.10 Define a function that labels the wines based on their quality according to the table below: <br>

|Quality Score|Label.|
|---:|---:|
|q>= 7|Top|
|5<q<7|Average|
|q<=5|Low|

```python
def quality(x):
    import numpy as np
    mustMeet = [(x['quality'] >= 7),(x['quality'] > 5) & (x['quality'] < 7), (x['quality']<=5)]
    result = ['Top', 'Average','Low']
    x['Qlabel'] = np.select(mustMeet, result)
    return x
```

### 2.11 Apply the function on the data frame and store the result in a new column "Qlabel".

```python
# your code here
quality(df)
```

### 2.12 Report the share of each quality label in percentage.

```python
# your code here
n1 = df[df['Qlabel']== 'Low']['Qlabel'].count()
n2 =  df[df['Qlabel']== 'Average']['Qlabel'].count()
n3 = df[df['Qlabel']== 'Top']['Qlabel'].count()
n4 = df['Qlabel'].count()

p = (n1/n4)*100
p1 = (n2/n4)*100
p2 = (n3/n4)*100

print('Percent of Low Quality Wine', p)
print('Percent of Average Quality Wine', p1)
print('Percent of Top Quality Wine', p2)
```
